```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Final_Project_PapaJohns/All_years.csv')
```

```python
df.columns
```

```python
# נוודא שסוג העמודות תקין
df["כמות"] = pd.to_numeric(df["כמות"], errors='coerce')
df["סכום"] = pd.to_numeric(df["סכום"], errors='coerce')

# סינון לפי כמות = 1 ובלי ערכים חסרים
filtered_df = df[(df["כמות"] == 1) & df["סכום"].notna() & df["תאור פריט"].notna()]

# מחשבים את הערך השכיח (mode) של סכום לכל תאור פריט
most_common_prices = (
    filtered_df.groupby("תאור פריט")["סכום"]
    .agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
)

# המרה למילון
price_dict = most_common_prices.to_dict()

# תצוגת דוגמה
for k, v in list(price_dict.items())[:10]:
    print(f"{k}: {v}")
```
