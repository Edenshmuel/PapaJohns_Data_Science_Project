```python
import pandas as pd
import numpy as np
```

```python
cleaned_data = pd.read_csv('/content/drive/MyDrive/Final_Project_PapaJohns/cleaned_data.csv')
```

```python
df = cleaned_data.copy()
```

```python
df.sample(3)
```

```python
# ערוצי מכירה והתפלגות הסתברויות (לפי גרף הפאי)
channels = ['Wolt', '10bis', 'PapaJohns App', 'Cibus', 'Telephone', 'Store']
probabilities = [0.37, 0.27, 0.18, 0.10, 0.05, 0.03]
```

```python
# קיבוע גרעין רנדום להגרלה שחוזרת על עצמה
np.random.seed(42)
```

```python
# הוספת עמודת ערוץ מכירה
df['sales_channel'] = np.random.choice(channels, size=len(df), p=probabilities)
```

```python
# הצגה לבדיקה
df.sample(3)
```

```python
# חישוב התפלגות כמותית
channel_counts = df['sales_channel'].value_counts()

# חישוב התפלגות באחוזים
channel_percentages = df['sales_channel'].value_counts(normalize=True) * 100

# שילוב שניהם בטבלה אחת
distribution_df = pd.DataFrame({
    'Count': channel_counts,
    'Percentage': channel_percentages.round(2)})
```

```python
# הצגה
print(distribution_df)
```
